```python
img_bgr = cv2.imread('/content/gdrive/MyDrive/Colab Notebooks/OpenCv/cv.jpg')
im_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(im_rgb)
plt.show()
```
